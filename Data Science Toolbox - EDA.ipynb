```python
df = pd.read_csv("downloads/train_eda.csv")
df = df.dropna(how ='any')
```

```python
le = LabelEncoder()

df.Embarked = le.fit_transform(df.Embarked)

df.head()
```

```python
cols = ['Pclass', 'Sex']
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
df.head()
```

- pd.cut() to create range and label for numerical to categorical group
- the bins will always be more than the label. i.e ( bins = 5 , label = 4)

```python
category = pd.cut(df.Age, bins=[0,2,17,65,99],labels=['Toddler/Baby','Child','Adult','Elderly'])
df.insert(5,'Age Group',category)
df.head()
```
